```python
#다음은 입력받은 숫자가 1000보다 크면 1자리부터 100자리까지의 숫자를 생략하고 ‘k’를 붙여주는 파이썬 코드입니다.

num = int(input())
result = str(num) 

if num >= 1000:
  result = str(num // 1000) + 'k'
elif num >=0:
  pass

print(result)
```

```python
sum = 0

while True: 
  num = int(input())
  if num < 0: 
      break 
  else: 
      sum = sum + num

print(sum)
```
